```python
import matplotlib.pyplot as plt

def read_tokenized_data(file_path):
    try:
        with open("tokenized_bengali.txt", 'r', encoding='utf-8') as file:
            content = file.read()
            tokens = content.split()
            return tokens
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found!")
        return []
    except Exception as e:
        print(f"Error reading file: {e}")
        return []
```

```python
file_path = "tokenized_bengali.txt"
print(f"Reading tokenized data from: {file_path}")
tokenized_data = read_tokenized_data(file_path)

if not tokenized_data:
    print("No data loaded. Please check your file path and format.")
    exit()

print(f"Successfully loaded {len(tokenized_data)} tokens from file")
```

- Counts the frequency of each token and returns a dictionary.

```python
def create_frequency_distribution(tokens):
    freq_dict = {}
    for token in tokens:
        if token in freq_dict:
            freq_dict[token] += 1
        else:
            freq_dict[token] = 1
    return freq_dict
```

- Plots a bar chart of the top N most frequent words.

```python
def plot_frequency_distribution(freq_dict, title, top_n=100):
    sorted_items = sorted(freq_dict.items(), key=lambda x: x[1], reverse=True)
    
    top_words = sorted_items[:top_n]
    words = [item[0] for item in top_words]
    frequencies = [item[1] for item in top_words]
    
    plt.figure(figsize=(15, 8))
    plt.bar(range(len(words)), frequencies)
    plt.title(title)
    plt.xlabel('Words (Top ' + str(len(words)) + ')')
    plt.ylabel('Frequency')
    
    plt.xticks(range(len(words)), words, rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
    return top_words
```

**Stopword identify**
- Returns a list of words whose frequency is greater than or equal to the threshold.

```python
def identify_stopwords(freq_dict, threshold):
    stopwords = []
    for word, freq in freq_dict.items():
        if freq >= threshold:
            stopwords.append(word)
    return stopwords
```

```python
def remove_stopwords(tokens, stopwords):
    filtered_tokens = []
    for token in tokens:
        if token not in stopwords:
            filtered_tokens.append(token)
    return filtered_tokens
```

```python
print("Creating frequency distribution...")
freq_dist = create_frequency_distribution(tokenized_data)

print(f"Total unique words: {len(freq_dist)}")
print(f"Total words: {len(tokenized_data)}")
```

```python
print("\nPlotting frequency distribution for top 100 words...")
top_100 = plot_frequency_distribution(freq_dist, "Top 100 Most Frequent Words", 100)
```

```python
thresholds = [3, 4, 5]

for threshold in thresholds:
    print(f"\n--- Analysis with threshold: {threshold} ---")
    
    stopwords = identify_stopwords(freq_dist, threshold)
    print(f"Stop words identified (freq >= {threshold}): {stopwords}")

    filtered_tokens = remove_stopwords(tokenized_data, stopwords)
    print(f"Tokens after removing stop words: {len(filtered_tokens)}")
    
    filtered_freq_dist = create_frequency_distribution(filtered_tokens)

    title = f"Top 100 Words After Removing Stop Words (Threshold >= {threshold})"
    plot_frequency_distribution(filtered_freq_dist, title, min(100, len(filtered_freq_dist)))
```
